```python
import yfinance 
import numpy as np # linear algebra
import pandas as pd # data
dateFrom = '2020-01-01'
dateTo = "2024-10-10"

tickers_list = ['YNDX', 'MSFT', 'AAPL', 'NVDA', 'MA']

# Import pandas
findata = pd.DataFrame(columns=tickers_list)

# Fetch the data

for ticker in tickers_list:
    loaded_data = yfinance.download(ticker, dateFrom, dateTo)
    findata[ticker] =loaded_data['Adj Close']

# Print first 5 rows of the data
findata.to_csv('findata.csv')
```
